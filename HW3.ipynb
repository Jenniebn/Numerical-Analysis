```python
# These are plots of Chebyshev polynomial for T1, T2, T3, and T6 by Jennie Wu
# Last Modified: Oct 13, 2021
# Input: The degree n and x value for Chebyshev polynomial
# Expected Output: Graph of Chebyshev polynomial for T1, T2, T3, and T6

import math
import matplotlib.pyplot as plt
import numpy as np

def T(n, x):
    if n == 0:
        return 1
    elif n == 1:
        return x
    elif n >= 1:
        return 2 * x * T(n - 1, x) - T(n - 2, x)
    
x = np.linspace(-1, 1, 101)
y1 = T(1, x)
y2 = T(2, x)
y3 = T(3, x)
y6 = T(6, x)
plt.axis("equal")
plt.plot(x, y1) #T1
plt.plot(x, y2) #T2
plt.plot(x, y3) #T3
plt.plot(x, y6) #T6
plt.grid()
plt.show()
```
